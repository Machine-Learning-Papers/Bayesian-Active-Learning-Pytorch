```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (12.0, 12.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

```

```python
data = {}
funs = ["VAR_RATIOS","MAX_ENTROPY", "MEAN_STD"]
limited_vr = funs #["VAR_RATIOS","RANDOM"]
train = {}
val = {}
test = {}
train_accuracy= {}
train_loss= {}
test_accuracy = {}
test_loss = {}
val_accuracy = {}
val_loss = {}
for x in funs:
    data[x] = {}
    for y in ["target","target_all","data_all","scores","scores_all"]:
        data[x][y] = np.load(x+"_Points_"+y+".npy")

pooled_points = data ["MAX_ENTROPY"]["scores_all"].argsort()[0:5][::-1]
# for image in data ["VAR_RATIOS"]["data_all"][pooled_points]:
# #     print image.shape
#     imgplot = plt.imshow(image[0])
    
# plt.show()
scores = data["MAX_ENTROPY"]["scores_all"][pooled_points]
for i, img in enumerate(data ["MAX_ENTROPY"]["data_all"][pooled_points]):
        
        plt.subplot(5, 10, i+1)
        plt.imshow(img[0])
        plt.axis('off')
        plt.title(str(scores[i])[0:5])
plt.show()
    
```

```python

'''
==============
3D scatterplot
==============

Demonstration of a basic scatterplot in 3D.
'''

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np


def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].

ax.scatter(data["VAR_RATIOS"]["target_all"], range(0,2000), data["VAR_RATIOS"]["scores_all"], c=data["VAR_RATIOS"]["scores_all"],cmap = "inferno", marker='o')

ax.set_xlabel('Target Labels')
ax.set_ylabel('Pool Subset Points')
ax.set_zlabel('Variational Ratios')

plt.show()
```
